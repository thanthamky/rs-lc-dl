```python
from sklearn.datasets import make_blobs, make_classification

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate a 3-class dataset with 3 features
X, y = make_classification(n_samples=100, n_features=3, n_informative=3, n_redundant=0, 
                           n_classes=3, weights=[0.5, 0.3, 0.2], random_state=42)

print("Class distribution:", np.bincount(y))
```

## Distribution and Counts

```python
select_feature = 0
select_class = 1

data = X[y==select_class][:, select_feature]

print(data)
print(f'Size: {len(data)}')
```

```python
# make widget

plt.figure(figsize=(6,4), dpi=120)

# make chart
plt.hist(data, 
         color='green'
         )

# make properties
plt.xlabel(f'feature {select_feature}')
plt.ylabel(f'class {select_class}')

plt.title('Histogram plot')

plt.grid()

plt.text(-2, 7, 'A Text on the plot', fontsize=12)

plt.show()
```

```python
plt.boxplot(data)
plt.ylabel('Feature 2')
plt.title('Box Plot of Feature 2')
plt.show()
```
